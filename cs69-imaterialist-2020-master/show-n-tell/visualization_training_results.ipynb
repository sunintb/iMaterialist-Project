# Visualization of Results for CS69.9/169.9 Team Project

```python
import numpy as np 
import pandas as pd
from pathlib import Path
import os
import warnings
import pickle

from matplotlib import pyplot as plt
import matplotlib.image as mpimg

warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
## Constants:
KAGGLE = False;
PREFIX = "/kaggle/" if KAGGLE else "../";

DATA_DIR = Path(PREFIX+'input/imaterialist-fashion-2020-fgvc7/')
IMG_DIR = Path(PREFIX+'input/imaterialist-fashion-2020-fgvc7/train/')
RESULTS_DIR = Path(PREFIX + 'input/');

## Load data:
df_train = pd.read_csv(DATA_DIR/'train.csv')
df_train.head()
```

```python
history = pickle.load(open(RESULTS_DIR/"saved_head_model_history.pkl",'rb')); 
epochs = range(0, 25);
```

```python
plt.figure(figsize=(6,6), dpi=100);
plt.plot(epochs, history["loss"], label="Training");
plt.plot(epochs, history["val_loss"], label="Validation");
plt.xlabel("Epoch", fontsize=15);
plt.ylabel("Loss", fontsize=15);
plt.legend(fontsize=15);
plt.show();
```
